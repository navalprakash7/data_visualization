### Functions of pyplot

```python
#all paramaneter of plt.plot() function, all these paramneter are optional except x,y
#plt.plot(x, y, color='red', linestyle='--', linewidth=2, marker='o', markersize=8, markerfacecolor='blue', markeredgecolor='black', label='Sales', alpha=0.8)
#how many cakes are sell in each day
days = ['Mon','Tues','Wed','Thus','Fri','Sat','Sun'] #x-axis (Horizontal)
sales = [10,7,90,50,30,22,15] #y-axis (Vertical)
plt.plot(days,sales,color='blue',linestyle='--',linewidth=2, marker='o',label='2026 Sales Data')
plt.title('Bakery Sales in this week') #Main title of graph
plt.xlabel('Days') #it will label on x axis
plt.ylabel('Cake sales per day') #it will label on y axis
plt.legend(loc='upper left',fontsize = 12) #this shows your lebel on graph
plt.grid(color='gray',linestyle=':',linewidth=1) #it will show grid lines
plt.xlim(0,3) #set x-axis range
plt.ylim(0,100) #set y-axis range
plt.xticks(['Mon','Tues','Wed','Thus','Fri','Sat','Sun'],['D1','D2','D3','D4','D5','D6','D7']) #it chnages "days name" to D1
plt.yticks([10,7,90,50,30,22,15],[1,7,0,5,3,2,4]) #it chnages the verticals numbers 
plt.show()
```

### Scatter plots, these plots mainly used in ML during clustring

```python
#plt.scatter(x, y, color='green', s=None, c=None, marker='o', cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, edgecolors=None, plotnonfinite=False, data=None, label=None)
hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]
plt.scatter(hours_studies, exam_scores, color='green',marker='o',label='Student Data')
plt.title('Trends')
plt.xlabel('hours_studies')
plt.ylabel('exam_scores')
plt.legend()
plt.show()
```

### Essentials subplot functions in matplotlib
#### plt.subplot() is used to create multiple plots in the same figure, so we can compare them easily

```python
#plt.subplot(nrows,ncols,index) 1 based indexing here
x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1) #1st row, 2nd col and 1st subplot
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2) #1st row, 2nd col and 2st subplot
plt.bar(x,y)
plt.title('Bar Graph')

plt.show()
```

```python
# fig, ax = plt.subplots(nrows, ncols, figsize=(width, height))
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

ax[0].plot(x, y, color='blue')
ax[0].set_title('Line Chart')

ax[1].bar(x, y, color='green')
ax[1].set_title('Bar Graph')

plt.suptitle('Comparisons of multiple graphs') #sets entire graphs title
plt.tight_layout() #adjusts the spacing between subplots and fix overlaps
plt.show()
```
