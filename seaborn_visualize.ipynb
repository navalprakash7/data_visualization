### ScatterPLot

```python
sns.set_style("ticks") #setting style or theme of your graph, chack different parameter_style from chatgpt
sns.set_context("talk") #it customize size of your grapsh using these paramenter (poster, talk, paper, notebook)
sns.scatterplot(data=df, x='flipper_length_mm', y='body_mass_g', hue='sex', palette="Set1") #hue is used to add a third variable using colors. and palette is for color customization
sns.despine(left=True) #it is used to remove a line from a graph
```

```python
sns.scatterplot(data=df, x='species', y='body_mass_g', hue='island', style="sex") #hue and style both are used to add extra variable but hue add based on color and style add based on shapes
```

### Strip Plot

```python
sns.stripplot(data=df, x='species', y='body_mass_g', hue='island', dodge=True, jitter=True)
```

###  Swarm Plot  

```python
sns.swarmplot(data=df, x='species', y='body_mass_g', hue='island')
```

###  Histogram  

```python
#in this graph we only analyse single variable, if you want add 1 more varible then use hue
sns.histplot(data=df, x='body_mass_g',hue='sex', multiple='stack')
```

###  Regression Plot

```python
sns.regplot(data=df, x='body_mass_g', y='flipper_length_mm', color='green')
```

```python
sns.regplot(data=df, x='body_mass_g', y='flipper_length_mm', color='green', scatter=False)
#scatter=False remove all scatter dots and only make line
```

###  Line Plot  

```python
sns.lineplot(data=df, x='body_mass_g', y='flipper_length_mm',hue='sex',style='island')
```

### Joint Plot

```python
# jointplot is used to visualize the relationship between two graphs
sns.jointplot(data=df, x='body_mass_g', y='flipper_length_mm', hue='sex') 
#if you dont put kind = 'type' it will choose other graph automatically
```

```python
sns.jointplot(data = df, x ="body_mass_g", y="flipper_length_mm", kind = "reg")
#now it shows reggresion plot and histrogram plot, also we only choose to decide only one type of plot
```

###  Bar Plot  

```python
sns.barplot(data = df, x='species', y ="body_mass_g")
```

```python
sns.barplot(data = df, x = "species", y ="body_mass_g", hue ="sex", palette =["pink", "blue"], estimator='sum') 
#if you didnot put estimator='sum' then default is mean also you can do more like, sum,mean,avr,standard dviation etc
```

### Count Plot  

```python
#this plot is used for counting vaule of column, just like pandas df.value_count() function
sns.countplot(data=df, x='species',hue='island', palette='Set2')
```

### Box Plot

```python
sns.boxplot(data = df, x="species", y ="body_mass_g", hue ="sex")
```

### Violin Plot  

```python
sns.violinplot(data = df, x="species", y ="body_mass_g", hue ="sex", split=True, inner='box')
```

```python
#violinplot & swarmplot in one plot
sns.violinplot(data = df, x="species", y ="body_mass_g")
sns.swarmplot(data = df, x="species", y ="body_mass_g", hue='island', size=3)
```

###  KDE Plot  

```python
#This plot is just like histogram but insted of bar it has smooth lines
sns.kdeplot(data = df, x ="body_mass_g", hue='species',fill=True)
```

###  Heatmap

```python

# Heatmap: a 2D color-coded matrix that shows relationships or patterns between two variables.
# It's commonly used to:
#- Show correlation between features
#- Visualize confusion matrices -> will see in the later stages.
#- Highlight missing data -> sns.heatmap(df.isnull() )
```

```python
columns = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
sns.heatmap(data = df[columns].corr(), annot=True, linewidth='2',linecolor='black')
```

###  Rug Plot 

```python
sns.rugplot(data = df, x = "body_mass_g", hue = "species", palette = "Set2", height=0.2)
```

###  Pair Plot

```python
#this is used for analysys all your columns at once
sns.pairplot(data = df, hue = "sex", palette = "Set2", diag_kind = "kde")
```

###  Pair Grid

```python
# pair grid -> subplot grid for plotting pairwise relationships in a dataset.
#it is similar to pair plot but we can mauanlly decide which graph show in upper, lower, digagonal trinangle
graph = sns.PairGrid(data = df, hue = "sex", palette = "Set2")
graph.map_upper(sns.scatterplot) #upper triangle
graph.map_lower(sns.kdeplot) #lower triangle
graph.map_diag(sns.histplot) #digonal of triangle
graph.add_legend()
```
